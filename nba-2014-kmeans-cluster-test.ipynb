```python
nba = pd.read_csv('2014-stats.csv')
```

```python
nba.loc[:, ].mean()
```

```python
kmeans_model = KMeans(n_clusters=5, random_state=1)
clean_model = nba._get_numeric_data().dropna(axis=1)
kmeans_model.fit(clean_model)
name = kmeans_model.labels_
name
```

```python
name1 = pd.DataFrame(name)
name1 = name1.rename(columns={0: "clusters"})
name2 = name1['clusters']
```

```python
name2
```

```python
clean_model['clusters'] = name2
clean_model
```

```python
Caron_Butler = clean_model.loc[nba['player'] == 'Caron Butler',: ]
Caron_Butler
```

```python
Caron_Butler_cluster = Caron_Butler.values.tolist()

kmeans_model.predict(Caron_Butler_cluster)
```
